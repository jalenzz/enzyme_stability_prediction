```python
import pandas as pd

train_features_df = pd.read_csv("../data/train_features.csv")
print(train_features_df.head())
print(train_features_df.info())
```

```python
# clip the data to 1000 rows
train_features_df = train_features_df.head(500)
# train_features_df.to_csv("../data/train_features_100.csv", index=False)
```

```python
import random
import matplotlib.pyplot as plt
# 分析数据
seq_feature = train_features_df["features"].values
seq_feature = [eval(i) for i in seq_feature]
tm = train_features_df["tm"].values

# 绘制特征图

random.seed(0)
# 10 random from 0 to 100
index = random.sample(range(100), 20)

for i in index:
    feature = [j[i] for j in seq_feature]
    fig, axs = plt.subplots(1, 2, figsize=(20, 5))
    
    # Plot feature
    axs[0].plot(feature)
    axs[0].set_title(f"{i}st Feature")
    
    # Plot feature vs tm
    axs[1].scatter(feature, tm)
    axs[1].set_title(f"{i}st Feature vs tm")
    
    plt.show()
```
